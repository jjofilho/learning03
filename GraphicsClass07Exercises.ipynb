```python
# Importa a biblioteca de gráficos
import matplotlib.pyplot as plt
```

```python
# Importa o arquivo excel para a análise como um dataframe
df = pd.read_excel('./recursos_recebidos.xlsx')
```

```python
# Mostra o dataframe
df
```

```python
# Cria o eixo X a partir de uma coluna da tabela em chaves, mas todos diferentes
eixo_x = df['LOCALIDADE'].keys()
```

```python
# Mostra os valores do eixo X
eixo_x
```

```python
# Cria o eixo X a partir de uma coluna da tabela
eixo_x = df['LOCALIDADE']
```

```python
# Mostra os valores do eixo X
eixo_x
```

```python
# Cria o eixo Y a partir de uma coluna da tabela
eixo_y = df['POPULAÇÃO']
```

```python
# Mostra os valores do eixo Y
eixo_y
```

```python
# Cria um gráfico em barras a partir dos eixos X e Y
plt.bar(eixo_x, eixo_y)
```

```python
# Cria um gráfico em barras a partir dos eixos X e Y, e inserção do título do gráfico
plt.bar(eixo_x, eixo_y)

plt.title('Distribuição da População por Município e DF')
```

```python
# Cria um gráfico em barras a partir dos eixos X e Y, inserção do título do gráfico e rótulos das barras com 90 graus no eixo X
plt.bar(eixo_x, eixo_y)

plt.title('Distribuição da População por Município e DF')

plt.xticks(rotation=90)
```

```python
# Cria um gráfico em barras a partir dos eixos X e Y
plt.bar(eixo_x, eixo_y)

# Insere o título do gráfico
plt.title('População Atendida com Recursos por Município e DF')

# Configura os rótulos das barras com 90 graus no eixo X
plt.xticks(rotation=90)

# Insere os rótulos dos eixos X e Y
plt.xlabel('Municípios')
plt.ylabel('População Atendida')
```

```python
# Cria uma coluna com o gasto médio por cidadão
media_gastos_cidadao = df.groupby(['LOCALIDADE'])['POPULAÇÃO'].agg(['mean']).reset_index()
```

```python
# Exibe a variável media_gastos_cidadao
media_gastos_cidadao
```

```python
# Organiza o groupby em ordem alfabética
media_gastos_cidadao = media_gastos_cidadao.sort_index()
```

```python
# Mostra o groupy organizado em ordem alfabética
media_gastos_cidadao
```

```python
# Atribuindo os valores dos eixos
localidade_x = media_gastos_cidadao['LOCALIDADE']
gastos_y = media_gastos_cidadao['mean']
```

```python
# Cria um gráfico em linha a partir dos eixos X e Y
plt.plot(localidade_x, gastos_y)

# Insere o título do gráfico
plt.title('Gasto Médio por Cidadão nos Municípios Atendidos')

# Configura os rótulos das barras com 90 graus no eixo X
plt.xticks(rotation=90)

# Insere os rótulos dos eixos X e Y
plt.xlabel('Municípios')
plt.ylabel('Média Gastos por Cidadão')
```

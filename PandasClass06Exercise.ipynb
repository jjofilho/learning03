```python
def limpa_valor(linha):
    linha['GASTOS DIRETOS'] = linha['GASTOS DIRETOS'].replace('R$ ', '')
    linha['GASTOS DIRETOS'] = linha['GASTOS DIRETOS'].replace('.', '')
    linha['GASTOS DIRETOS'] = linha['GASTOS DIRETOS'].replace(',', '')
    valores = int(linha['GASTOS DIRETOS'])
    linha['GASTOS DIRETOS'] = valores
    return linha
```
